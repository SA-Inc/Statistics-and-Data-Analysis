```python
import numpy as np
from scipy import stats
import math
# import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Исходные данные
dataSet = np.array([3,6,8,11,6,10,7,9,7,3,4,8,2,7,9,4,9,11,7,8,4,10,5,6,7])
```

```python
print(dataSet)
```

```python
# Частоты значений в датасете
unique, counts = np.unique(dataSet, return_counts=True)
frequency = np.asarray((unique, counts))
```

```python
print(frequency)
```

```python
fig, ax = plt.subplots()
ax.plot(frequency[0], frequency[1], label="Частота")
ax.set_title('Полигон частот')
ax.set_xlabel('Зачения')
ax.set_ylabel('Частота')
ax.legend()
plt.xticks(np.arange(min(frequency[0]), max(frequency[0]) + 1, 1.0))
plt.grid()
plt.show()
```

```python
# Средневзвешенная частот
mean = np.average(frequency[0], weights = frequency[1])
variance = np.average((frequency[0] - mean) ** 2, weights = frequency[1])
std = math.sqrt(variance)
median = np.median(dataSet)
mode = unique[counts == counts.max()][0]
```

```python
print(mean)
print(variance)
print(std)
print(median)
print(mode)
```

```python
fig, ax = plt.subplots()
ax.plot(range(len(dataSet)), dataSet, label = "Значения")
ax.set_title('Data Set')
# ax.set_xlabel('')
ax.set_ylabel('Зачения')
ax.axhline(y = mean, color = "red", label = "Средневзвешенное")
ax.axhline(y = median, color = "green", label = "Медиана")
ax.axhline(y = mode, color = "yellow", label = "Мода")
# ax.axhline(y = variance, label = "Дисперсия")
# ax.axhline(y = std, label = "Стандартное отклонение")
ax.legend()
plt.xticks(np.arange(0, len(dataSet)))
plt.yticks(np.arange(min(dataSet), max(dataSet)))
plt.grid()
plt.show()
```
